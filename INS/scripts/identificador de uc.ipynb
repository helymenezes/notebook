```python
import pandas as pd
#Leitura de pastas
path_cadastro  = r'E:\OEM\content\cadastro.xlsx'
path_concluido= r'E:\OEM\content\base_sip_Concluido.xlsx'
path_eac = r'E:\INST\content\MLA_EAC_Ligações_Intelbras.23052024.xlsx'
#Leitura do arquivo excel
read_cad = pd.read_excel(path_cadastro)
read_concluido = pd.read_excel(path_concluido)
read_eac = pd.read_excel(path_eac)
#Selecionando as colunas
df_cad =  read_cad[['UC', 'IDSIGFI','OBRA','ROTA']]
df_concluido = read_concluido[['UC', 'IDSIGFI','OBRA','ROTA']]
df_eac = read_eac [['Uc', 'Código ONE','Nº Obra','ROTA']]
```

```python
select_2t = [14, 15, 16, 17, 21, 22, 23, 24]
cad = df_cad.loc[df_cad['ROTA'].isin(select_2t)]
concluido = df_concluido.loc[df_concluido['ROTA'].isin(select_2t)]
eac = df_eac.loc[df_eac['ROTA'].isin(select_2t)]
df_concat = pd.concat([cad,concluido], axis=0)
df_date = eac.merge(df_concat,left_on= 'Código ONE', right_on= 'IDSIGFI', how='left')#'IDSIGFI'
date = df_date[df_date['UC'].isna()]
date = date.drop_duplicates()
```
